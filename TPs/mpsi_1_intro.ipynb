!!! info Nommer ses variables

En python, les noms de variables doivent commencer par une lettre ou un underscore (`_`), et peuvent ensuite contenir des lettres ou des underscore. \
En mathématiques, on a tendance à utiliser des noms de variables le plus court possible : `n`, `k`, ...\
En informatique, c'est l'inverse ! On préfère avoir des noms de variables explicites, pour que le code soit plus clair. Visez un (ou éventuellement deux) mots pour la majorité de vos variables.

En Python, la convention est d'utiliser la `snake_case` pour le nom des variables. Cela signifie qu'on ne met pas de majuscules (sauf pour les acronymes), et que les mots sont séparés par des underscore `_`. Je vous demande de respecter cette convention.
!!!


Les valeurs (et donc les variables associées) ont également des _types_. Par exemple, la variable `a` contient un nombre entier. Le type correspondant est `int`.

Les types de base de python sont les suivants : 
- `int` pour les entiers de n'importe quelle taille

## Les instructions conditionnelles
Les instructions conditionelles permettent de d'exécuter du code différent en fonction d'une condition. 

La syntaxe est la suivante : 
```
if condition:
    #faire quelque chose
elif condition: 
    #faire quelque chose
else: 
    #faire quelque chose
```

La première instruction, `if` ("si" en anglais), execute son bloc si la condition est vérifié. \
On peut ensuite avoir un nombre quelconque (0 ou plus) instructions `elif`, qui s'executent si : 
- les conditions précédentes n'ont pas été vérifiées
- la condition du `elif` est vérifiée
Enfin, on peut avoir (optionel) un bloc `else`, qui s'execute si aucune des conditions précédentes n'ont été vérifiées.

Les conditions sont des booléens : on peut écrire `false` directement, mais ça n'a pas beaucoup d'interêt. La plupart du temps, la condition sera 
- soit une variable qui contient un booléen
- soit une comparaisons entre deux valeurs

!!! warning
Attention, n'oubliez pas les `:` !
!!!

!!! info Indentation
En Python, l'indentation est obligatoire. Cela signifie que les blocs de votre `if` doivent être plus indenté (donc plus à droite, ou avec plus d'espace au début) que les instructions `if`..`elif`..`else`

## Boucles

Les boucles sont des constructions du language qui permettent de ... Boucler. C'est-à-dire, de répéter des instructions. Il existe deux types de boucles : 
- les boucles bornées, qui permettent de boucler un nombre prédéfini de fois
- les boucles non bornées, qui bouclent tant qu'une condition n'est pas vérifiée

### Boucles bornées

Les boucles bornées s'écrivent "`for`... `in` ...", qui signifique "Pour chaque ... dans ...". La construction la plus commune est la suivante :


````python
for i in range(10):
    print(i)
````

Ici, le bloc interne se répète **10** fois, et la variable `i` prend toutes les valeurs de 0 à (10 - 1).

!!! info Itération 
On appelle *itération* un tour de boucle. Ici, on a eu 10 itérations.

!!! info La fonction `range`
`range` permet de générer une suite (finie) d'entiers. On peut soit lui donner un argument n, comme dans l'exemple ci-dessous, pour avoir une suite allant de 0 à n-1. 

Mais on peut en fait lui donner trois arguments avec `range(start, end, step)`. Dans ce cas, elle donne la suite d'entiers qui : 
- Commence à `start` (inclus)
- Progresse de `step` en `step` (par exemple, si `step = 2`, la suite sera `start, start + 2, start + 4, ...`
- S'arrête avant de dépasser `end` (exclus)

### Boucle non bornées
Les boucles non bornées s'écrivent avec le mot-clé `while` (qui signifie "tant que"), suivi d'une condition (comme le mot-clé `if`). 

````python
#ce code n'est volontairement pas commenté
seuil = 1000
n_iter = 0
i = 1
while seuil >= 0:
    seuil -= i
    i += 1
    n_iter += 1
print(n_iter)
````

Attention, si vous mettez une condition qui est toujours vérifiée sur une boucle while, votre programme bouclera à l'infini. On dit qu'il ne termine pas. Ce n'est à priori jamais souhaitable d'avoir un programme qui ne termine pas, donc vous devrez toujours vous assurer que cela n'arrive pas.

!!! info Quelle boucle utiliser ? 
La boucle `for` est plus restrictives que la boucle `while`, mais est plus facile à utiliser (il n'y a pas le risque de non-terminaison). Ainsi, si vous pouvez utiliser une boucle `for`, faites-le, et sinon, utilisez la boucle `while`.

## Les fonctions

Vous avez déjà utilisé des fonctions : `range` et `print` sont des fonctions dites "de la librairie standard", c'est-à-dire qu'elles sont toujours disponible.

On peut également définir nos propres fonctions. Une fonction, c'est un bout de code réutilisable. Utiliser des fonctions a plusieurs avantages : 
- Pas besoin de réécrire le code, il suffit d'appeler la fonction par son nom
- Cela donne un programme mieux organisé : chaque fonctionnalité est placée dans sa "boite" (la fonction)
- Même si ce n'est pas forcément pertinent pour vous, cela permet aussi de partager son code, pour qu'il soit utilisables par d'autres

En fait, la majorité du code que vous écrivez sera placé dans des fonctions, et, par la suite, la plupart des questions et exercices vous demenderont d'écrire une fonction qui réalise un certain objectif. 

Pour définir une fonction, on utilise le mot clé `def`, suivi d'un nom de fonction (en *snake_case*, comme les variables), suivi d'une liste d'arguments entre parenthèses.

Une fonction peut *renvoyer* un résultat avec le mot-clé `return`, ce qui interromp immédiatement son execution.

!!! info Documentation
On peut utiliser le triple guillement pour ajouter de la documentation (une forme de commentaire) à notre fonction.
Toutes vos fonctions doivent être documentés avec les informations suivantes :
- En une ligne, que fait la fonction (si ce n'est pas évident)
- Liste des arguments et leur type
- Type de retour
!!!

````python
def add(x, y):
    """
    Ajoute deux entiers
    Arguments : x et y deux entiers
    Résultat : leur somme
    """
    return x + y
````
