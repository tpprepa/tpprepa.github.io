## 1. Motifs et égalité 

!!! info Info
L'instruction `grille = BlockGrid(12, 12, fill = (240, 240, 240))` permet de créer une grille carrée de 12 cellules de côté. La couleur de chaque cellule est un gris clair de code RGB(240, 240, 240).

```python
grille = BlockGrid(12, 12, fill = (240, 240, 240))
grille.show()
```


 
Chaque cellule possède les attributs :

* row : indice de ligne de la cellule
* col : indice de colonne de la cellule

  
Au survol de la souris, les coordonnées `[row, col]` et la couleur de la cellule sont affichées.

!!!

!!! question Exercice 1 question 1
Exécuter la cellule ci-dessous et tester le survol de la souris pour repérer les coordonnées des cellules
!!!

````python
grille = BlockGrid(12, 12, fill = (240, 240, 240))
grille.show()
````

!!! info Info
La méthode `set_colors()` permet de modifier les niveaux de (rouge, vert, bleu). Pour modifier la couleur de la cellule située à la ligne 0 et à la colonne 2 et la mettre en rouge,  on écrit le code : 
```python
grille[0, 2].set_colors(255, 0, 0) 
```

ou simplement
```python
grille[0, 2] = (255, 0, 0)
```
Pour voir les modifications, il faut à nouveau écrire l'instruction `grille.show()` pour voir le changement.

2. Créer votre propre motif en créant et combinant les fonctions de votre choix.

````python
grille = BlockGrid(12, 12, fill = (240, 240, 240))

def motif_19():
for cellule in grille:
    if (cellule.row // 2 + cellule.col * 3) % 7 == 0 :
        cellule.set_colors(0, 0, 0)

def motif_20():
for cellule in grille:
    if (cellule.row * 3 + cellule.col // 2) % 7 == 0 :
        cellule.set_colors(0, 0, 0)

motif_19()
motif_20()
grille.show()
````
