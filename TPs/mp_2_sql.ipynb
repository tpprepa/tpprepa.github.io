Ce TP vise à exploiter une base de donnée réelle grâve au language SQL. Normalement, vous devriez pouvoir afficher la liste des tables disponibles dans la base de donnée en exécutant la cellule suivante : 

````sql
.tables
````

Il s'agit d'une base de donnée nationale sur les élèves dans l'enseignement supérieur en France, pour l'année 2023. Les données sont bien sûr anonymisées. Les attributs des trois tables sont : 



!!! abstract **Communes** 
Contient les communes de France qui contiennent un établissement. 
- clé primaire `code` : le code postal de la commune 
- `nom` : le nom de la commune 
- `departement` : le département de la commune 
- `region` : la région de la commune
!!!

!!! abstract **Etablissements**
Contient (une partie) des établissements d'enseignement supérieur en France.
- clé primaire `id` : un identifiant d'établissement
- `type` : le type d'établissement 
- `nom` : le nom de l'établissement
- clé étrangère `code_commune` : le code postal de la commune du siège de l'établissement. 
!!!

!!! abstract **Eleves**
Contient des groupes d'élèves correspondant à une catégorie, ainsi que leur nombre (avec et sans CPGE)/
- clé étrangère `id_etablissement` : l'identifiant de l'établissement
- `sexe` : "M" ou "F", l'état civil de(s) éléves (malheureusement entièrement binaire en France)
- `serie_bac` : le bac obtenu
- `departement_bac` : si le bac a été passé dans un autre département que celui de l'établissement ou non.
- `diplome` : le diplôme préparé
- `discipline` : la discipline étudiée. À noter que seule une partie des disciplines est présente ici (de la Chimie aux Maths globalement). 
- `nb_etudiants_sans_CPGE` : nombre d'étudiants dans le groupe, sans compter ceux en double inscription CPGE
- `nb_etudiants_avec_CPGE` : idem mais avec les doubles inscriptions.

La clé primaire de cette table est (`id_etablissement`, `sexe`, `serie_bac`, `departement_bac`, `diplome`, `discipline`)
!!!
