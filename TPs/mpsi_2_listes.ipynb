!!! info Courbes en python
Il est possible d'afficher des courbes en python à l'aide de la librairie `matplotlib.pyplot`.
Pour cela, il commencer par *importer* la librairie, avec la ligne : 
```python
import matplotlib.pyplot as plt
```

Pour afficher une courbe, il faut ensuite utiliser les trois instructions suivantes : 
```python 
plt.clf()
plt.plot(X, Y, "r-")
plt.show()
```
- la première instruction remet à zéro les courbes qui auraient pu être tracées précédemment
- la deuxième instruction créé la courbe elle-même. Elle prend en argument :
    * deux listes ou tableaux (ici `X` et `Y`), qui doivent être de la même taille
    * une chaine de caractère pour indiquer la forme : ici, "r+" signifie une ligne rouge. La première lettre est la couleur, et le deuxième charactère est la forme des points (par exemple, il existe `+`, `*`, `.`, ...)

Une expression parfois utile pour afficher des courbes est `list(range(n))` qui permet de construire la liste des entiers de 0 à n-1 (En fait, il s'agit simplement de convertir la séquence renvoyée par `range` en liste).

!!! question Question 16
Je vous fourni une fonction `chrono(f, arg1, arg2, ...)` qui mesure le temps d'exécution d'une fonction. Utilisez-là pour comparer les deux fonctions de calcul de nombre premiers. Si vous le souhaitez, tracez une courbe du temps d'exécution des deux fonctions par rapport à `n`.
!!!

````python
# Vous n'avez pas à comprendre le code de cette cellule, simplement à l'exécuter.
import time
def chrono(f, *args):
    """
    Prend en argument une fonction `f` et ses arguments. 
    La fonction chrono affiche ensuite le temps d'execution de la fonction `f`, et renvoie la valeur de retour de `f`.
    """
    before = time.time()
    ret_value = f(*args)
    total_time = time.time() - before
    print(f"La fonction s'est executée en {total_time} secondes.")
    return ret_value
````

````python
# Démonstration de la fonction chrono
def attends(secondes):
    """
    Prend en argument un flottant. Attends ce nombre de secondes, puis termine.
    """
    time.sleep(secondes)

chrono(attends, 1.5)
````
